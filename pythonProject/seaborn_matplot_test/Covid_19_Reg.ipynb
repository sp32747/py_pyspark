```python
maha=df[df.state=='Maharashtra']
maha
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x='date',y='confirmed',data=maha,color='red')
plt.show()
```

```python
from datetime import datetime as dt
```

```python

maha['date']=maha['date'].map(dt.toordinal)
maha.tail(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=maha['date']
y=maha['death']
```

```python
 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrm=LinearRegression()
```

```python
lrm.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
lrm.score(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```
